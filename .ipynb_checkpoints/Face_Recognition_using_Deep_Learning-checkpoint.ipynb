```python
import numpy as np
def get_label(image_name):
    name = image_name.split('.')[-3]
    if(name=='Gaurav'):
        return np.array([1,0,0])
    if(name=='Paati'):
        return np.array([0,1,0])
    if(name=='Uday'):
        return np.array([0,0,1])
```

```python
import os
from random import shuffle
from tqdm import tqdm
```

```python
import cv2
def my_data():
    data = []
    for img in tqdm(os.listdir("data")):
        path = os.path.join("data", img)
        img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img_data = cv2.resize(img_data, (50,50))
        data.append([np.array(img_data), get_label(img)])
    shuffle(data)
    return data
```

```python
data = my_data()
```

```python
import cv2
import numpy as np

def get_predict(model):
    # Load the face classifier
    face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    
    def croped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        if len(faces) == 0:
            return None, None
        # Assuming there's only one face in the frame, you might want to handle multiple faces differently
        (x, y, w, h) = faces[0]
        crop_img = gray[y:y+h, x:x+w]
        return crop_img, (x, y, w, h)

    cap = cv2.VideoCapture(0)

    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        face, bbox = croped(frame)
        if face is not None:
            face = cv2.resize(face, (50, 50))
            X = np.array(face).reshape(1, -1)  # Reshape to (1, 2500) for prediction
            prediction = model.predict(X)
            label = ["Gaurav", "Paati", "Uday"]
            print(f"Predicted class label: {label[prediction[0]]}")

            # Draw rectangle around detected face
            (x, y, w, h) = bbox
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            # Put prediction text on the frame
            
            cv2.putText(frame, f"Prediction: {label[prediction[0]]}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

        cv2.imshow('Face', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Assuming you have already trained your SVM model and stored it in 'model'
# Call the function with your model
get_predict(model)
```
